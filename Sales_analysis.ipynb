```python
import pandas as pd
import sqlite3 as sql
import matplotlib.pyplot as plt
```

```python
connection = sql.connect('sales_data.db')
```

```python
query = '''SELECT product_name AS product, SUM(quantity) AS total_quantity, SUM(quantity * unit_price) AS revenue
 FROM sales
 GROUP BY product_name
 '''
```

```python
data_df = pd.read_sql_query(query, connection)
```

```python
print(data_df)
```

```python
data_df.plot(kind = 'bar', x = 'product', y = 'revenue')
```

```python
plt.savefig('sales_chart.png')
```
